```python
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=["c","c#","php","react.js"]
print("pie plot")
plt.pie(x,labels=y)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[2,4,5,1]
print("scatter plot")
plt.scatter(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[2,4,5,7]
y=[3,4,1,6]
plt.title("stem plot")
plt.stem(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[1,2,3,4]
plt.title("step plot")
plt.step(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,3,5,6]
area=[2,5,1,4]
area1=[3,2,2,4]
plt.title("stackplot")
plt.stackplot(x,area,area1)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=[12,24,36,48]
plt.title("histograph")
plt.hist(x)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=[2,4,6,7]
plt.plot(x,color="red")
plt.xlabel("language")
plt.ylabel("code")
plt.title("line plot")
plt.show()
```
